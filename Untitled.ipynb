# TIME SERIES ANALYSIS 

```python
df.columns
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')

print(df['Order Date'].isna().sum())


df = df.dropna(subset=['Order Date'])

df.set_index('Order Date', inplace=True)
```

```python
monthly_sales = df['Sales'].resample('ME').sum()
```

```python
plt.figure(figsize=(12,6))
plt.plot(monthly_sales, marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

# Customer and Product Analysis

```python
customer_sales = df.groupby('City')['Sales'].sum()
customer_sales.plot(kind='bar')
plt.title('Sales by Customer Demographics')
plt.xlabel('Customer Demographic')
plt.ylabel('Sales')
plt.show()
```

```python
product_sales = df.groupby('Category')['Sales'].sum()
product_sales.plot(kind='bar')
plt.title('Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.show()
```

```python
df.columns = df.columns.str.strip()
df['Order Date'] = pd.to_datetime(df['Order Date'].str.strip(), errors='coerce')
df = df.dropna(subset=['Order Date'])  # Drop rows where 'Order Date' could not be converted

# Set 'Order Date' as the index
df.set_index('Order Date', inplace=True)

# Resample the data to monthly sales for smoother trends
monthly_sales = df['Sales'].resample('ME').sum()

# Pivot data for heatmap
monthly_sales_df = monthly_sales.reset_index()
monthly_sales_df['Year'] = monthly_sales_df['Order Date'].dt.year
monthly_sales_df['Month'] = monthly_sales_df['Order Date'].dt.month

# Pivot table for heatmap
heatmap_data = monthly_sales_df.pivot_table(values='Sales', index='Year', columns='Month', fill_value=0)

# Create heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap='coolwarm', cbar_kws={'label': 'Sales'})
plt.title('Monthly Sales Heatmap')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```
